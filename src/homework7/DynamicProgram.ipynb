```python
# EX 1:
# Khai báo cấu trúc cuộc họp
class Meeting:
    def __init__(self, a, b, num):
        self.a = a  # Thời điểm bắt đầu
        self.b = b  # Thời điểm kết thúc
        self.num = num  # Số thứ tự của cuộc họp

# Hàm so sánh để sắp xếp
def compare(x):
    return (x.a, x.b)

# Hàm chính
def max_meetings(n, meetings):
    # Sắp xếp danh sách cuộc họp theo thời gian bắt đầu, sau đó là thời gian kết thúc
    meetings.sort(key=compare)

    # Mảng f lưu trữ số lượng cuộc họp tối đa không chồng lấn tại mỗi điểm
    # Mảng d lưu trữ chỉ số cuộc họp trước đó để truy vết
    f = [0] * (n + 1)
    d = [0] * (n + 1)

    # Bước quy hoạch động để tìm số lượng cuộc họp tối đa
    for i in range(1, n + 1):
        f[i] = 1  # Tại mỗi vị trí, ít nhất có thể chọn cuộc họp hiện tại
        for j in range(1, i):
            if meetings[j - 1].b <= meetings[i - 1].a and f[i] < f[j] + 1:
                f[i] = f[j] + 1
                d[i] = j

    # Tìm chỉ số cuộc họp cuối cùng có số lượng lớn nhất
    t = 1
    for i in range(2, n + 1):
        if f[i] > f[t]:
            t = i

    # Truy vết chuỗi cuộc họp đã chọn
    seq = []
    while t:
        seq.append(meetings[t - 1].num)
        t = d[t]

    # Trả về chuỗi cuộc họp theo thứ tự đã chọn
    return list(reversed(seq))

# Ví dụ đầu vào
n = int(input("Nhập số lượng cuộc họp: "))  # Nhập số lượng cuộc họp
meetings = []
for i in range(1, n + 1):
    a, b = map(int, input(f"Nhập thời gian bắt đầu và kết thúc cuộc họp {i}: ").split())
    meetings.append(Meeting(a, b, i))

# Tìm và hiển thị các cuộc họp đã chọn
# n = 6
# meetings = [(1, 3), (2, 4), (3, 5), (7, 8), (6, 10), (5, 7)]
selected_meetings = max_meetings(n, meetings)
print("Các cuộc họp được chọn:", ' '.join(map(str, selected_meetings)))
```

```python
# EX2:
# Danh sách các yêu cầu với thời gian bắt đầu, thời gian kết thúc và giá thuê
requests = [
    (1, 3, 50),
    (2, 5, 20),
    (4, 6, 60),
    (6, 7, 30),
    (5, 8, 90),
    (8, 9, 40),
]

# Sắp xếp các yêu cầu theo thời gian kết thúc
requests.sort(key=lambda x: x[1])

def find_previous(requests, index):
    for i in range(index - 1, -1, -1):
        if requests[i][1] <= requests[index][0]:
            return i
    return -1

# Hàm tính lợi ích tối đa và truy vết các yêu cầu được chọn
def maximize_profit(requests):
    n = len(requests)
    # Lợi ích tối đa cho từng yêu cầu
    dp = [0] * n
    selected_requests = [None] * n

    # Lợi ích tối đa cho yêu cầu đầu tiên là chính giá trị của nó
    dp[0] = requests[0][2]
    selected_requests[0] = [requests[0]]

    for i in range(1, n):
        # Tìm yêu cầu trước đó mà không chồng chéo
        previous = find_previous(requests, i)

        # Lợi ích khi chọn yêu cầu này
        current_profit = requests[i][2]

        # Nếu có yêu cầu không chồng chéo trước đó
        if previous != -1:
            current_profit += dp[previous]

        # Lợi ích tối đa khi chọn hoặc không chọn yêu cầu này
        if current_profit > dp[i - 1]:
            dp[i] = current_profit
            # Kết hợp yêu cầu hiện tại với các yêu cầu trước đó nếu chọn
            if previous != -1:
                selected_requests[i] = selected_requests[previous] + [requests[i]]
            else:
                selected_requests[i] = [requests[i]]
        else:
            dp[i] = dp[i - 1]
            # Nếu không chọn, lấy danh sách yêu cầu của trước đó
            selected_requests[i] = selected_requests[i - 1]

    # Tổng lợi ích tối đa
    max_profit = dp[-1]

    # Danh sách yêu cầu được chọn
    best_selection = selected_requests[-1]

    return max_profit, best_selection

# Tính lợi ích tối đa và danh sách các yêu cầu được chọn
max_profit, selected_requests = maximize_profit(requests)

# Hiển thị kết quả
print("Lợi ích tối đa:", max_profit)
print("Danh sách các yêu cầu được chọn:", selected_requests)
```

```python
# EX3:
import math

# Hàm tính diện tích của tam giác dựa trên tọa độ 3 đỉnh
def triangle_area(x1, y1, x2, y2, x3, y3):
    return abs((x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)) / 2)

# Hàm kiểm tra xem điểm (px, py) có nằm trong tam giác (x1, y1, x2, y2, x3, y3) không
def is_point_in_triangle(px, py, x1, y1, x2, y2, x3, y3):
    area_main = triangle_area(x1, y1, x2, y2, x3, y3)
    area_1 = triangle_area(px, py, x2, y2, x3, y3)
    area_2 = triangle_area(x1, y1, px, py, x3, y3)
    area_3 = triangle_area(x1, y1, x2, y2, px, py)
    return math.isclose(area_main, area_1 + area_2 + area_3)

# Hàm kiểm tra xem tam giác t1 có bao tam giác t2 không
def does_triangle_contain(t1, t2):
    return (
        is_point_in_triangle(
            t2[0], t2[1], t1[0], t1[1], t1[2], t1[3], t1[4], t1[5]
        )
        and is_point_in_triangle(
            t2[2], t2[3], t1[0], t1[1], t1[2], t1[3], t1[4], t1[5]
        )
        and is_point_in_triangle(
            t2[4], t2[5], t1[0], t1[1], t1[2], t1[3], t1[4], t1[5]
        )
    )

# Hàm tìm dãy dài nhất của các tam giác bao nhau
def longest_contained_triangles(triangles):
    # Sắp xếp tam giác theo diện tích tăng dần
    triangles.sort(key=lambda t: triangle_area(t[0], t[1], t[2], t[3], t[4], t[5]))
    
    # Quy hoạch động để tìm dãy dài nhất
    n = len(triangles)
    dp = [1] * n  # Khởi tạo mỗi tam giác có thể tạo thành dãy độ dài 1
    
    for i in range(1, n):
        for j in range(0, i):
            if does_triangle_contain(triangles[i], triangles[j]):
                dp[i] = max(dp[i], dp[j] + 1)
    
    return max(dp)

# Ví dụ đầu vào
triangles = [
    [0, 0, 1, 0, 0, 1],  # Tam giác 1
    [0, 0, 0.5, 0.5, 0, 0.5],  # Tam giác 2
    [0, 0, 2, 0, 0, 2],  # Tam giác 3
]

# Tìm dãy tam giác bao nhau dài nhất
result = longest_contained_triangles(triangles)

print("Dãy tam giác bao nhau dài nhất:", result)
```

```python
# EX4:
# Khai báo các biến cần thiết
N = 2003
a = [0] * N
P = [0] * N
Q = [0] * N

# Đầu vào cố định
n = 6  # Số lượng phần tử
U = 10  # Giá trị U
L = 2  # Giá trị L
a = [0, 3, 8, 12, 15, 20, 25]  # Mảng a với phần tử đầu tiên là 0 (vì chỉ số bắt đầu từ 1)

# Tính toán các giá trị P và Q
for i in range(1, n + 1):
    P[i] = 1
    Q[i] = 1
    for j in range(1, i - L):
        if a[i] - U <= a[j] < a[i]:  # Điều kiện cho Q
            Q[i] = max(Q[i], P[j] + 1)
        if a[i] < a[j] <= a[i] + U:  # Điều kiện cho P
            P[i] = max(P[i], Q[j] + 1)

# Tìm giá trị lớn nhất
mx = Q[1]
for i in range(1, n + 1):
    mx = max(mx, max(P[i], Q[i]))

# In kết quả
print("Giá trị lớn nhất:", mx)
```

```python
# EX5:
# Hàm tính dãy con tăng dần dài nhất kết thúc tại mỗi vị trí
def longest_increasing_sequence_ending_at(a):
    n = len(a)
    Q = [1] * n
    for i in range(1, n):
        for j in range(i):
            if a[j] < a[i]:
                Q[i] = max(Q[i], Q[j] + 1)
    return Q

# Hàm tính dãy con giảm dần dài nhất bắt đầu từ mỗi vị trí
def longest_decreasing_sequence_starting_from(a):
    n = len(a)
    P = [1] * n
    for i in range(n - 2, -1, -1):
        for j in range(n - 1, i, -1):
            if a[j] < a[i]:
                P[i] = max(P[i], P[j] + 1)
    return P

# Hàm tìm dãy con WAVIO có độ dài lớn nhất
def longest_wavio_subsequence(a):
    n = len(a)
    # Tính các dãy con tăng dần kết thúc tại mỗi vị trí
    Q = longest_increasing_sequence_ending_at(a)
    # Tính các dãy con giảm dần bắt đầu từ mỗi vị trí
    P = longest_decreasing_sequence_starting_from(a)

    # Tìm giá trị lớn nhất của Q[i] + P[i] - 1
    max_wavio = 0
    for i in range(n):
        max_wavio = max(max_wavio, Q[i] + P[i] - 1)

    return max_wavio

# Ví dụ sử dụng
a = [1, 2, 3, 4, 5, 2, 1]  # Dãy ví dụ
result = longest_wavio_subsequence(a)

print("Độ dài dãy WAVIO lớn nhất:", result)
```

```python
# EX 6:
# Khối đá với các chiều dài, rộng, cao
class Stone:
    def __init__(self, dimensions):
        self.dimensions = sorted(dimensions)  # Sắp xếp để chiều cao là kích thước nhỏ nhất
        self.height = self.dimensions[0]  # Chiều cao là kích thước nhỏ nhất
        self.length = self.dimensions[1]  # Chiều dài là kích thước lớn thứ hai
        self.width = self.dimensions[2]  # Chiều rộng là kích thước lớn nhất

# Hàm chính để tìm số lượng khối đá nhiều nhất có thể xếp chồng
def max_stone_tower(stones):
    n = len(stones)
    
    # Tạo một danh sách các khối đá với tất cả các khả năng xoay
    all_stones = []
    for stone in stones:
        # Tạo các cách xoay khác nhau
        a, b, c = stone
        all_stones.append(Stone((a, b, c)))
        all_stones.append(Stone((a, c, b)))
        all_stones.append(Stone((b, c, a)))
    
    # Sắp xếp các khối đá theo chiều dài và sau đó là chiều rộng
    all_stones.sort(key=lambda x: (x.length, x.width))

    # Quy hoạch động để tìm số lượng khối đá nhiều nhất
    dp = [1] * len(all_stones)

    # Kiểm tra từng khối đá xem chúng có thể xếp chồng lên nhau không
    for i in range(1, len(all_stones)):
        for j in range(0, i):
            if (all_stones[j].length < all_stones[i].length and
                all_stones[j].width < all_stones[i].width):
                dp[i] = max(dp[i], dp[j] + 1)
    
    # Số lượng lớn nhất của khối đá có thể được xếp chồng lên nhau
    return max(dp)

# Ví dụ đầu vào
stones = [(3, 4, 1), (8, 6, 2), (7, 3, 1)]

# Kết quả: tìm số lượng khối đá nhiều nhất có thể xếp chồng
result = max_stone_tower(stones)

print("Số lượng khối đá nhiều nhất có thể xếp chồng:", result)
```

```python
# EX 7:
# Hàm kiểm tra xem có thể tạo ra tổng S từ dãy con A hay không
def can_create_subset_sum(A, S):
    n = len(A)

    # Khởi tạo mảng L với S + 1 phần tử, mặc định là 0
    L = [0] * (S + 1)
    # Dãy con rỗng luôn có tổng 0
    L[0] = 1

    # Quy hoạch động để cập nhật khả năng tạo ra tổng
    for i in range(n):
        # Đi ngược từ S về giá trị của phần tử hiện tại
        for t in range(S, A[i] - 1, -1):
            if L[t] == 0 and L[t - A[i]] == 1:
                L[t] = 1

    # Kiểm tra xem tổng S có thể được tạo ra hay không
    return L[S] == 1

# Ví dụ đầu vào
A = [3, 34, 4, 12, 5, 2]
S = 9

# Kiểm tra xem tổng S có thể được tạo ra từ dãy con của A không
if can_create_subset_sum(A, S):
    print("Có dãy con có tổng bằng S")
else:
    print("Không có dãy con có tổng bằng S")
```

```python
# EX8:
# Số lượng gói kẹo
n = 6

# Số lượng kẹo trong từng gói
a = [2, 4, 5, 3, 9, 8]

# Tổng số lượng kẹo
t = sum(a)

# Tạo ma trận quy hoạch động với kích thước (n + 1) x (t // 2 + 1)
L = [[False for _ in range(t // 2 + 1)] for _ in range(n + 1)]

# Tổng bằng 0 luôn có thể tạo ra từ dãy con rỗng
for i in range(n + 1):
    L[i][0] = True

# Quy hoạch động để xác định nếu tổng t có thể tạo ra từ các phần tử A[1] đến A[i]
for i in range(1, n + 1):
    for j in range(1, t // 2 + 1):
        L[i][j] = L[i - 1][j]  # Sử dụng kết quả từ gói trước đó
        if a[i - 1] <= j:
            L[i][j] = L[i][j] or L[i - 1][j - a[i - 1]]  # Xem xét phần tử hiện tại

# Tìm giá trị lớn nhất của j sao cho L[n][j] là True và 2 * j <= t
mx = 0
for i in range(1, t // 2 + 1):
    if L[n][i]:
        mx = i

# Chênh lệch giữa hai phần
difference = t - 2 * mx

# In ra các kết quả
print("Số lượng kẹo trong phần đầu tiên:", mx)
print("Số lượng kẹo trong phần thứ hai:", t - mx)
print("Chênh lệch giữa hai phần:", difference)
```

```python
# EX 9:
def find_possible_weights(fish_weights):
    # Tính tổng khối lượng tất cả cá
    total_weight = sum(fish_weights)

    # Mảng đánh dấu các tổng khả thi
    L = [False] * (total_weight + 1)
    # Tổng bằng 0 luôn khả thi
    L[0] = True

    # Cập nhật mảng đánh dấu để tính các tổng khả thi
    for weight in fish_weights:
        # Đi ngược lại để tránh ghi đè tổng khả thi trước đó
        for t in range(total_weight, weight - 1, -1):
            if L[t - weight]:
                L[t] = True

    # Tìm các tổng khả thi
    possible_weights = [t for t in range(1, total_weight + 1) if L[t]]

    return possible_weights

# Ví dụ sử dụng
fish_weights = [3, 2, 4]  # Khối lượng cá
possible_weights = find_possible_weights(fish_weights)

print("Các tổng khả thi là:", possible_weights)
```

```python
# EX10:
def solve(A, S):
    n = len(A)
    # Tính tổng của các số
    total = sum(A)
    # Khởi tạo mảng L có kích thước (n+1) x (2*total+1) và gán giá trị False cho tất cả phần tử
    L = [[False] * (2 * total + 1) for _ in range(n + 1)]
    # Gán True cho trường hợp đặc biệt khi không có số nào và tổng S là 0
    L[0][total] = True
    
    # Duyệt qua từng số và cập nhật giá trị cho mảng L
    for i in range(1, n + 1):
        for t in range(2 * total + 1):
            if L[i - 1][t]:
                # Nếu có thể đặt dấu "+" hoặc "-" sau khi thêm số thứ i để có tổng t
                if t + A[i - 1] <= 2 * total:
                    L[i][t + A[i - 1]] = True
                if t - A[i - 1] >= 0:
                    L[i][t - A[i - 1]] = True
    
    # Kiểm tra xem có cách nào để đạt được tổng S không
    if S + total >= 0 and S + total <= 2 * total and L[n][S + total]:
        return True
    else:
        return False

# Example usage:
A = [1, 3, 5, 7]
S = 6
if solve(A, S):
    print("Có cách để đạt được tổng S")
else:
    print("Không có cách để đạt được tổng S")
    
# Example usage:
A = [1, 3, 4, 7]
S = 6
if solve(A, S):
    print("Có cách để đạt được tổng S")
else:
    print("Không có cách để đạt được tổng S")
```

```python
def max_product_sum_partition(nums):
    total = sum(nums)
    max_product = 0
    
    # Tạo mảng đánh dấu để lưu trữ các tổng có thể đạt được
    possible_sums = [False] * (total + 1)
    possible_sums[0] = True
    
    current_sum = 0
    for num in nums:
        # Cập nhật các tổng có thể đạt được sau khi thêm số num vào
        for i in range(total, -1, -1):
            if possible_sums[i]:
                possible_sums[i + num] = True
        current_sum += num
    
    # Tạo danh sách chứa các số thuộc nhóm thứ nhất
    group1 = []
    s_max = 0
    for s in range(total + 1):
        if possible_sums[s]:
            if s * (total - s) > max_product:
                max_product = s * (total - s)
                s_max = s
    
    # Xây dựng nhóm thứ nhất
    current_sum = 0
    for num in reversed(nums):
        if current_sum + num <= s_max:
            group1.append(num)
            current_sum += num
            nums.remove(num)
    
    # In ra các số thuộc nhóm thứ hai
    print("Nhóm 1:", group1)
    print("Nhóm 2:", nums)
    print("Tích của tổng 2 nhóm lớn nhất là:", max_product)

# Example usage:
nums = [1, 2, 3, 4, 5]
print("Cách chia dãy số sao cho tích của tổng 2 nhóm là lớn nhất:")
max_product_sum_partition(nums)
```

```python
# EX12:
def max_olive_trees(N, M, A, B, Q):
    # Khởi tạo mảng 2 chiều để lưu giá trị tối đa có thể đạt được
    # Mỗi hàng của mảng này đại diện cho việc xét thêm một mảnh đất hoặc một dải đất
    # Mỗi cột của mảng này đại diện cho khối lượng (số cây ôliu) tối đa có thể đạt được
    dp = [[0] * (Q + 1) for _ in range(N + M + 1)]

    # Duyệt qua từng mảnh đất và dải đất
    for i in range(1, N + M + 1):
        if i <= N:  # Xét mảnh đất
            W = A[i - 1]  # Khối lượng (số cây ôliu) của mảnh đất
            V = A[i - 1]  # Giá trị của mảnh đất cũng chính là khối lượng
        else:  # Xét dải đất
            W = B[i - N - 1]  # Khối lượng (số cây ôliu) của dải đất
            V = B[i - N - 1]  # Giá trị của dải đất là số cây ôliu của mảnh đất trước đó

        # Duyệt qua từng khối lượng có thể đạt được
        for j in range(Q + 1):
            # Nếu khối lượng của đồ vật hiện tại nhỏ hơn hoặc bằng khối lượng có thể đạt được
            if W <= j:
                # Lưu giá trị tối đa có thể đạt được
                dp[i][j] = max(dp[i - 1][j], dp[i - 1][j - W] + V)
            else:
                # Nếu không, giá trị tối đa giữ nguyên
                dp[i][j] = dp[i - 1][j]

    # Trả về giá trị tối đa có thể đạt được
    return dp[N + M][Q]

# Example usage:
N = 3
M = 2
A = [2, 3, 4]
B = [1, 2]
Q = 5
print("Số lượng cây ôliu tối đa có thể trồng:", max_olive_trees(N, M, A, B, Q))
```
