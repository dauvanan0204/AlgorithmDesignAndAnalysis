```python
## Biểu đồ fibonacci
# sizeOfData = {1, 10, 20, 30, 40, 45}
# Time no recursive: [300, 500, 400, 500, 700, 800]
# Time recursive: [300, 6400, 694500, 6417000, 522567000, 5740346600]
import matplotlib as matplotlib
import matplotlib.pyplot as plt

x = [1, 10, 20, 30, 40, 45]
no_recursive = [300, 500, 400, 500, 700, 800]
recursive = [300, 6400, 694500, 6417000, 522567000, 5740346600]

plt.plot(x, no_recursive, 'b-', label = "Non Recursive")
plt.plot(x, recursive, color = 'r', linestyle = '-', label = "Recursive")

plt.legend() # Chu thich
plt.xlabel('Number of element') 
plt.ylabel('Time to run(nns)')
plt.show()
```
