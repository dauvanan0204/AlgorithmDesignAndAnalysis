```python
import random

# Ma trận chi phí (giả sử là một đồ thị đầy đủ)
cost_matrix = [
    [0, 3, 10, 25, 6, 18, 27, 22, 24],
    [3, 0, 11, 3, 14, 12, 7, 24, 22],
    [10, 11, 0, 8, 27, 13, 24, 17, 18],
    [25, 3, 8, 0, 24, 11, 9, 23, 16],
    [6, 14, 27, 24, 0, 9, 15, 10, 21],
    [18, 12, 13, 11, 9, 0, 14, 29, 30],
    [27, 7, 24, 9, 15, 14, 0, 6, 17],
    [22, 24, 17, 23, 10, 29, 6, 0, 19],
    [24, 22, 18, 16, 21, 30, 17, 19, 0]
]

# Số lượng quần thể
population_size = 20

# Số lần lặp tối đa
max_generations = 1000

# Tạo một cá thể ngẫu nhiên
def create_individual():
    return random.sample(range(len(cost_matrix)), len(cost_matrix))

# Tính toán chi phí của một cá thể (đường đi)
def calculate_cost(individual):
    cost = 0
    for i in range(len(individual) - 1):
        cost += cost_matrix[individual[i]][individual[i+1]]
    cost += cost_matrix[individual[-1]][individual[0]]
    return cost

# Lựa chọn cha mẹ từ quần thể
def select_parents(population):
    return random.sample(population, 2)

# Lai ghép (crossover)
def crossover(parent1, parent2):
    crossover_point = random.randint(0, len(parent1) - 1)
    child1 = parent1[:crossover_point] + [gene for gene in parent2 if gene not in parent1[:crossover_point]]
    child2 = parent2[:crossover_point] + [gene for gene in parent1 if gene not in parent2[:crossover_point]]
    return child1, child2

# Đột biến (mutation)
def mutate(individual, mutation_rate):
    for i in range(len(individual)):
        if random.random() < mutation_rate:
            j = random.randint(0, len(individual) - 1)
            individual[i], individual[j] = individual[j], individual[i]
    return individual

# Tiến hành thuật toán di truyền
def genetic_algorithm():
    # Khởi tạo quần thể ban đầu
    population = [create_individual() for _ in range(population_size)]

    # Lặp qua các thế hệ
    for generation in range(max_generations):
        # Lựa chọn và lai ghép
        new_population = []
        for _ in range(population_size // 2):
            parent1, parent2 = select_parents(population)
            child1, child2 = crossover(parent1, parent2)
            new_population.extend([child1, child2])

        # Đột biến
        for individual in new_population:
            mutate(individual, mutation_rate=0.1)

        # Chọn lọc
        population = sorted(new_population, key=lambda x: calculate_cost(x))[:population_size]

        # In thông tin về thế hệ hiện tại
        best_individual = population[0]
        best_cost = calculate_cost(best_individual)
        print(f"Generation {generation+1}: Best Cost = {best_cost}, Best Route = {best_individual}")

    return population[0], calculate_cost(population[0])

# Chạy thuật toán di truyền
best_route, best_cost = genetic_algorithm()
print(f"Best Route: {best_route}")
print(f"Best Cost: {best_cost}")
```
